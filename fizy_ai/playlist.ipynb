```python
  # OYNATMA LISTESINE GORE ONE CIKAN muzik turu, sanatci ve muzik donemini saptayan fonksiyon tanimlama
def top(pl1, df):
    ta = list(pl1['Artist'].value_counts().index[:4])    # top artists:  oynatma listesindeki en populer sanatci
    ratio_pop = (df[df['Genre'] == "pop"].count() / df["Song"].count())[0]
    ratio_hiphop = (df[df['Genre'] == "hip hop"].count() / df["Song"].count())[0]
    ratio_rock = (df[df['Genre'] == "rock"].count() / df["Song"].count())[0]
    ratio_folk = (df[df['Genre'] == "folk"].count() / df["Song"].count())[0]
    ratio_blues = (df[df['Genre'] == "blues"].count() / df["Song"].count())[0]
    
    ratios = {'pop': ratio_pop, 'hiphop': ratio_hiphop, 'rock': ratio_rock, 'folk': ratio_folk, 'blues': ratio_blues}
    #print(ratio_pop+ ratio_hiphop+ ratio_rock+ ratio_folk+ ratio_blues)
    
      # OYNATMA LISTESINDEKI MUZIK TURU ORANI
    pl1_r = {'pop': (pl1[pl1['Genre'] == "pop"].count() / pl1["Song"].count())[0], 
             'hiphop': (pl1[pl1['Genre'] == "hip hop"].count() / pl1["Song"].count())[0], 
             'rock': (pl1[pl1['Genre'] == "rock"].count() / pl1["Song"].count())[0], 
             'folk': (pl1[pl1['Genre'] == "folk"].count() / pl1["Song"].count())[0], 
            'blues': (pl1[pl1['Genre'] == "blues"].count() / pl1["Song"].count())[0]}
    
    
      # OYNATMA LISTESINDE DINLENMIS BUTUN MUZIKLERE NAZARAN HANGI MUZIK TURUNUN AGIR BASTIGINI SAPTAMA
    d = {}
    for i in pl1_r:
        d.update({i: (pl1_r[i]-ratios[i])*10})
    
    tg = heapq.nlargest(2, d.keys(), key=lambda k: d[k])  # top genres:  oynatma listesindeki en populer tur
      # GENEL MUZIK donemi ORANLARI
    ratio_70 = (df[df['Decade'] == "70s"].count() / df["Song"].count())[0]
    ratio_80 = (df[df['Decade'] == "80s"].count() / df["Song"].count())[0]
    ratio_90 = (df[df['Decade'] == "90s"].count() / df["Song"].count())[0]
    ratio_00 = (df[df['Decade'] == "00s"].count() / df["Song"].count())[0]
    ratio_10 = (df[df['Decade'] == "10s"].count() / df["Song"].count())[0]
    ratio_20 = (df[df['Decade'] == "20s"].count() / df["Song"].count())[0]
    ratio_old = (df[df['Decade'] == "old"].count() / df["Song"].count())[0]
    
    ratios = {'70s': ratio_70, '80s': ratio_80, '90s': ratio_90, '00s': ratio_00, '10s': ratio_10, '20s': ratio_20, 'old': ratio_old}
    
    
      # OYNATMA LISTESINDEKI MUZIK donemi ORANI
    pl1_r = {'70s': (pl1[pl1['Decade'] == "70s"].count() / pl1["Song"].count())[0], 
             '80s': (pl1[pl1['Decade'] == "80s"].count() / pl1["Song"].count())[0], 
             '90s': (pl1[pl1['Decade'] == "90s"].count() / pl1["Song"].count())[0], 
             '00s': (pl1[pl1['Decade'] == "00s"].count() / pl1["Song"].count())[0], 
             '10s': (pl1[pl1['Decade'] == "10s"].count() / pl1["Song"].count())[0], 
             '20s': (pl1[pl1['Decade'] == "20s"].count() / pl1["Song"].count())[0], 
             'old': (pl1[pl1['Decade'] == "old"].count() / pl1["Song"].count())[0], }
    
    
      # OYNATMA LISTESINDE DINLENMIS BUTUN MUZIKLERE NAZARAN HANGI 10 yillik donemin AGIR BASTIGINI SAPTAMA
    dd = {}
    for i in pl1_r:
        dd.update({i: (pl1_r[i]-ratios[i])*10})
    
    td = heapq.nlargest(2, dd.keys(), key=lambda k: dd[k])  # top decade:  oynatma listesindeki en populer donem
    return {'artists': ta, 'genres': tg, 'decades': td}

```
